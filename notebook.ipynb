```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "data/athlete_events.csv"

df = pd.read_csv(filepath)
df.shape
```

```python
nulls = df.isnull().sum().drop("Medal")

plt.figure(figsize=(5,3))
nulls.plot(kind="bar", color="mediumseagreen")
plt.show()
```
