```python
import pandas as pd
import json
```

```python
df = pd.read_csv('data/polished3.csv')
```

```python
df.describe()
```

```python
# for each row in the csv create a property in the json object with their unique items, ignore height, weight, bmi, year, for those get the max and the min

#df = df.drop(columns=["ID", "Unnamed: 0","Name"])
dict_to_json = {}
# for each columns
for column in df.columns:
    if column == "ID" or column == "Unnamed: 0":
        continue
    if column == "Height" or column == "Weight" or column == "BMI" or column == "Year" or column == "Age":
        dict_to_json[column] = {"min": df[column].min(), "max": df[column].max()}
    else:
        dict_to_json[column] = df[column].unique().tolist()

dict_to_json
```

```python
import numpy as np


def convert_to_serializable(obj):
    if isinstance(obj, (np.integer, np.floating)):
        return obj.item()  # Convert NumPy numbers to native Python types
    elif isinstance(obj, np.ndarray):
        return obj.tolist()  # Convert NumPy arrays to lists
    raise TypeError(f"Type {type(obj)} is not serializable")

json_object = json.dumps(dict_to_json, indent = 4, default=convert_to_serializable)

with open("data/polished3.json", "w") as outfile:
    outfile.write(json_object)
```
