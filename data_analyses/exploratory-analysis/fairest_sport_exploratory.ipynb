```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
base_file_path = '../../data/'


df = pd.read_csv(base_file_path+"polished.csv")
```

```python
df["BMI"] = df["Weight"] / (df["Height"] / 100) ** 2
df["BMI"].head()
sns.boxplot(data=df, y ="BMI", hue="Sex")
```

```python
bmi_by_sport = df.groupby("Sport")["BMI"].mean().sort_values()
plot = sns.barplot(bmi_by_sport)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
df["Won Medal"] = df.Medal != "No Medal"
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Height'], df['Won Medal'])
print(slope, intercept, r_value, p_value, std_err)
```

```python
slopes = df.groupby('Sport').apply(lambda group: stats.linregress(group['Height'], group['Won Medal'])[0])

slopes_df = slopes.reset_index(name='LC').sort_values(by="LC")
slopes_df
```

```python
plot = sns.barplot(data=slopes_df, x="Sport", y="LC")
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
male = df[df.Sex == "M"]
female = df[df.Sex == "F"]

slopes_male = male.groupby('Sport').apply(lambda group: stats.linregress(group['Height'], group['Won Medal'])[0])

slopes_male_df = slopes_male.reset_index(name='LC').sort_values(by="LC")

slopes_female = female.groupby('Sport').apply(lambda group: stats.linregress(group['Height'], group['Won Medal'])[0])

slopes_female_df = slopes_female.reset_index(name='LC').sort_values(by="LC")
```

```python
plot = sns.barplot(data=slopes_male_df, x="Sport", y="LC")
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
plot = sns.barplot(data=slopes_female_df, x="Sport", y="LC")
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
slope_bmi = df.groupby('Sport').apply(lambda group: stats.linregress(group['BMI'], group['Won Medal'])[0])

slopes_bmi_df = slope_bmi.reset_index(name='LC').sort_values(by="LC")
slopes_bmi_df
```

```python
slope_age = df.groupby('Sport').apply(lambda group: stats.linregress(group['Age'], group['Won Medal'])[0])

slope_age_df = slope_age.reset_index(name='LC').sort_values(by="LC")
slope_age_df
```

```python
medal_rate = df.groupby('Sport').apply(lambda df: df[df["Won Medal"] == True].shape[0] / df.shape[0]).sort_values()
medal_rate
```

```python
plot = sns.barplot(medal_rate)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```
